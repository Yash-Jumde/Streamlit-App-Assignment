```python
import pandas as pd
```

```python
file_path = "./Task3/rawdata.xlsx"

df = pd.read_excel(file_path)
df.head()
```

```python
df['date_str'] = df['date'].astype(str)
df['time_str'] = df['time'].astype(str)

df
```

```python
df['datetime'] = pd.to_datetime(df['date_str'] + ' ' + df['time_str'])
df.drop(columns=['date_str', 'time_str'], inplace=True)
df['position'] = df['position'].str.lower()
```

```python
df
```

```python
df['duration'] = df['datetime'].diff().fillna(pd.Timedelta(seconds=0))

duration_data = df.groupby(['date', 'position'])['duration'].sum().reset_index()

activity_data = df.groupby(['date', 'activity']).size().reset_index(name='count')

combined_df = pd.merge(duration_data, activity_data, on=['date'], how='outer')
```

```python
combined_df
```

```python
combined_df.to_csv('combined_data.csv', index=False)
```
